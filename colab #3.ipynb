```python
import numpy as np
```

```python
pi = 3.14
t = np.arange(0,30)  # 구간 정수로 나눠짐
# t = np.arange(0,30,0.01)  # = 구간을 0.01 단위로 나누겠다
y = np.cos((pi/3)*t) +  np.sin((pi/2)*t)

plt.plot(t,y)
plt.show()
```

```python
t = np.arange(0, 10, 0.01)
x1 = np.heaviside(t-2,1)  # numpy 에서 step function을 그리는 함수
x2 = np.heaviside(t-3,1)  # (시작위치, 길이)
x3 = np.heaviside(t-5,1)

x = x1+x2-x3  # 다양한 조합 가능

plt.plot(t,x)
plt.show()
```
